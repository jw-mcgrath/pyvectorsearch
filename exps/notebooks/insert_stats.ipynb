```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("../output_data/insert_stats.csv")
```

```python
data.head()
```

```python


plt.figure()

# Plot the data
plt.plot(data['size'], data['visited'], color='blue')

# Add titles and labels
plt.title('Traversals as a function of size')
plt.xlabel('Graph size (nodes)')
plt.ylabel('Travserals per insert')
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure()

# Plot the data
plt.plot(data['size'], data['search_distance_computations'], color='blue')

# Add titles and labels
plt.title('distance computations as a function of size')
plt.xlabel('Graph size (nodes)')
plt.ylabel('distance computations')
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
