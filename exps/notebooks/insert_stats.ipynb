```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_c = pd.read_csv("../output_data/insert_stats_collated.csv")
data_nc = pd.read_csv("../output_data/insert_stats_non_collated.csv")
```

```python
data_c.head()
```

```python


plt.figure()

# Plot the data
plt.plot(data_c['size'], data_c['visited'], label='collated', color='blue')
plt.plot(data_nc['size'], data_nc['visited'], label='non collated', color='red')

# Add titles and labels
plt.title('Traversals as a function of size')
plt.xlabel('Graph size (nodes)')
plt.ylabel('Travserals per insert')
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure()

# Plot the data
plt.plot(data_c['size'], data_c['search_distance_computations'],label='collated', color='blue')
plt.plot(data_nc['size'], data_nc['search_distance_computations'],label='non collated', color='red')
# Add titles and labels
plt.title('distance computations as a function of size')
plt.xlabel('Graph size (nodes)')
plt.ylabel('distance computations')
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
